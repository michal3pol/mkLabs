```python
import numpy as np
import pandas as pd

data = pd.read_csv("war4.csv", delimiter=";", decimal=",", header=0)

print(data)
```

```python
x1 = data["x1"].values.astype(float)
x2 = data["x2"].values.astype(float)
y = data["y"].values

# A matrix
A = np.column_stack((x1, x2))
U,S,VT = np.linalg.svd(A, full_matrices=False)
print("U")
print(U)
print("VT")
print(VT)
```

```python
a,b = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ y

print("Obliczone współczynniki: a = " + str(a) + ", b = " + str(b))
```
